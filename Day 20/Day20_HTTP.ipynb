Fetching contents from web page

```python
import requests

def fetch_webpage_content_requests(url):
    try:
        response = requests.get(url)
        response.raise_for_status() 
        return response.text
    except requests.exceptions.RequestException as e:
        return f"Error fetching URL: {e}"

if __name__ == '__main__':
    url = "https://www.geeksforgeeks.org/python/python-network-programming/" 
    content = fetch_webpage_content_requests(url)
    print(content)
```
